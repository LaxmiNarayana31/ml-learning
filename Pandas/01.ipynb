```python
# Creating a DataFrame
df = pd.DataFrame([[1,2,3], [4,5,6], [7.8,9]])
print("Without Labels :\n", df)

# Also we can provide labels for these data
df_with_labels = pd.DataFrame([[1,2,3], [4,5,6], [7.8,9]], columns=['Col1', 'Col2', 'Col3'], index=['Row1', 'Row2', 'Row3'])
print("With Labels :\n", df_with_labels)
```

```python
# Accessing data
dataset.loc[0, 'Age'] # Accessing Age of first row using label-based indexing
```

```python
# Group of rows and columns
dataset.loc[0:4, 'Age']  # Accessing Age of first 5 rows
dataset.loc[0:4, ['Age', 'Weight']]  # Accessing Age and Weight of first 5 rows
```

```python
# Range of rows and all columns
"""Means from row index 0 to 4 (inclusive) and all columns from 'Age' to 'FAVC' (inclusive)."""
dataset.loc[0:4, "Age":"FAVC"]  
```

```python
# To see all rows from a specific column range
dataset.loc[:, 'Age':'FAVC']
```

```python
# To see the specific rows and specific columns
dataset.loc[50:70, 'Age':'FAVC']
```

```python
# in iloc we use integer-based indexing
# dataset.iloc[row_range, column_range]
dataset.iloc[0:10, 0:5]
```

```python
# If we want to single value access for that we use at and iat. Slicing is not allowed here.
# at and iat expects both row and column labels/indices to be provided as single values.
# dataset.at[row_label, column_label]
dataset.at[0, "Age"]
```

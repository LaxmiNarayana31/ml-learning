```python
dataset_path = r"C:\Users\Sumit Amit\OneDrive\Desktop\ML_Notebook\dataset\obesity_dataset.csv"
dataset = pd.read_csv(dataset_path)

# Remove all spaces around column names
dataset.columns = dataset.columns.str.strip()
dataset.columns
```

```python
# Same thing using lambda function
dataset["Height"] = dataset["Height"].apply(lambda height: height * 100)
dataset.head(7)
```

```python
# Calculate BMI and add as a new column
def calculate_bmi(row):
    height_m = row["Height"] / 100
    weight_kg = row["Weight"]
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 2)
```

```python
# Add BMI column to dataset
dataset["BMI"] = dataset.apply(calculate_bmi, axis=1)
dataset.head(10)
```

```python
# Add a new column in a soecific position
bmi_column = dataset.pop("BMI")
# in insert it expect the index position, column name and the data
dataset.insert(4, "BMI", bmi_column)
dataset.head(10)
```

```python
# To drop any column from the dataset
# If you does not specify inplace=True it will return a new dataframe without modifying the original one or you can assign it back to the original variable
# if you want to drop multiple columns you can pass a list of column names to the columns parameter
dataset.drop(columns=["BMI"], inplace=True)
dataset.head(10)
```

```python
# Again add BMI column to dataset
dataset["BMI"] = dataset.apply(calculate_bmi, axis=1)
dataset.head(10)
```

```python
# Add a new column in a soecific position
bmi_column = dataset.pop("BMI")
# in insert it expect the index position, column name and the data
dataset.insert(4, "BMI", bmi_column)
dataset.head(10)
```

```python
# Or you can do this in a single line
del dataset["BMI"]
dataset.head(10)
```

```python
# Rename columns
dataset.rename(columns={"Weight": "Weight_kg", "Height": "Height_cm"}, inplace=True)
dataset.head(10)
```
